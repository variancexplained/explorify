# T-Test

```python
import pandas as pd
from explorify.eda.stats.inferential.centrality import TTest
from studioai.visual.container import AnalyzerContainer
```

## Dependencies

```python
container = AnalyzerContainer()
container.init_resources()
container.wire(modules=["explorify.eda.stats.inferential.centrality"])
```

## Data

```python
FP = "data/Credit Score Classification Dataset.csv"
df = pd.read_csv(FP, index_col=None)
male = df.loc[df["Gender"] == "Male", "Income"]
female = df.loc[df["Gender"] == "Female", "Income"]
```

```python
test = TTest(a=male, b_name=female, varname="Income", a_name="Male", b_name='Female')
test.run()
test.result.plot()
```
