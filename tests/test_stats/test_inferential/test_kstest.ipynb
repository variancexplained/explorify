# KS Test

```python
import pandas as pd
import numpy as np
from explorify.eda.stats.inferential.gof import KSTest
from studioai.visual.container import AnalysisContainer
```

## Dependencies

```python
container = AnalysisContainer()
container.init_resources()
container.wire(modules=["explorify.eda.stats.inferential.gof"])
```

## Data

```python
FP = "data/Credit Score Classification Dataset.csv"
df = pd.read_csv(FP, index_col=None)
income = df['Income'].values
age = df['Age'].values
```

```python
test = KSTest(a=income, b_name="norm")
test.run()
test.result.plot()
```

```python
test = KSTest(a=income, b_name=age)
test.run()
test.result.plot()
```

```python
a = np.random.normal(size=1000)
test = KSTest(a=a, b_name="norm")
test.run()
test.result.plot()
```
