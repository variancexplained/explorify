# Apps per Category

```python
import pandas as pd
from explorify.utils.io import IOService
```

```python
FP_EXP = "tests/data/apps_by_category.csv"
FP_REVIEWS = "tests/data/reviews.pkl"
FP_APPS_BY_CATEGORY = "tests/data/apps_by_category.pkl"
```

```python
fp_exp = IOService.read(FP_EXP)[["Category", "Apps"]]
fp_exp.columns = ["Category", "Expected"]
fp_exp
```

```python
reviews = IOService.read(FP_REVIEWS)
fp_actual = reviews.groupby(by="category")["app_name"].nunique().reset_index()
fp_actual.columns = ["Category", "Actual"]
fp_actual
```

```python
df = pd.merge(fp_exp, fp_actual, on="Category")
IOService.write(filepath=FP_APPS_BY_CATEGORY, data=df)
```
